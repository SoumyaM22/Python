```python
#Subplot means inside 1 chart if there are multiple sub plots
import matplotlib.pyplot as plt

x =[1,2,3,4,5] 
y = [45,34,56,23,45]

plt.subplot(1,2,1) #subplot(row,column,chartnumber)
plt.title("age")
plt.suptitle("employee data") #Supertitle
plt.plot(x,y)
plt.show()

x = [5,6,7,8,9]
y = [67,50,66,56,82]

plt.subplot(1,2,2) #subplot(row,column,index)
plt.title("weight")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [45, 34, 56, 23, 45]
plt.subplot(2,2,1) #rows, columns, chartnum 
plt.plot(x,y)

x = [5, 6, 7, 8, 9]
y = [67, 58, 66, 56, 82]
plt.subplot(2,2,2)
plt.bar(x,y)

x = [2, 4, 6, 8, 10]
y = [57, 50, 60, 55, 20]
plt.subplot(2,2,3)
plt.scatter(x,y)

x = [1, 3, 5, 7, 9]
y = [67, 51, 62, 58, 23]
plt.subplot(2,2,4)
plt.step(x,y)
plt.show()
```
