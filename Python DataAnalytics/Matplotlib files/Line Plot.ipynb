```python
import matplotlib.pyplot as plt
x = ["day1","day2","day4","day4","day5"]
y = [300,420,250,230,400]
y1 = [500,450,300,250,320]
#plt.plot(x,y,marker="o")
#plt.plot(x,y,marker="*")
#plt.plot(x,y,marker="d")
#plt.plot(x,y,marker="v")
#plt.plot(x,y,marker="v",ls = ":")
plt.figure(figsize = [5,3]) # to reduce figure size
plt.plot(x,y,marker="^",ls = "--",color="red",label="week1")
plt.plot(x,y1,marker="*",ls = "-",color="green",label="week2",alpha=0.5)
# linestyle or ls: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
# alpha value is (0 to 1) we use to lighten color of line
plt.legend() # to show label info
plt.show()
```
