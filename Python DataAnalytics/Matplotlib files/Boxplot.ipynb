```python
#Box Plot is used for statistical analytics
"""Boxplots provide a visual summary of key data characteristics:

Median: The central value of the dataset.
Quartiles: The 25th percentile (Q1) and 75th percentile (Q3), which define the interquartile range (IQR).
Whiskers: Indicate the range of the data, typically extending to 1.5 * IQR beyond the quartiles.
Outliers: Data points outside the whiskers, which are plotted individually."""

"""
Video 14:39- diagram
Min- Lower fence
    Min/Lower fence = Q1 - 1.5(IQR) 
Q1- 25th percentile(25%)
  =(25/100)*8+1 (Here 8- no. of values -[1,3,4,7,12,2,8,9,24])
Median
Q3- 75th percentile
  =(75/100)*9
Max- Higher fence
    Max/Upper fence = Q3 + 1.5(IQR) 

IQR - Distance from Q1 to Q3
IQR = Q3 - Q1

Eg1: list - [1,3,4,7,12,2,8,9,24]
Q1 =(25/100)*10=2.5
Q3 =(75/100)*10=7.5
IQR = 7.5-2.5=5
Min/Lower fence = Q1 - 1.5(IQR)=2.5 - 1.5(5)=2.5-7.5=-5
Max/Upper fence = Q3 + 1.5(IQR) = 7.5 + 1.5(5)=7.5 +7.5 = 15
In list, nearby min value to -5 is 1
In list, nearby max value to 15 is 12
ascending order of list- [1,2,3,4,7,8,9,12,24]
median=7, for 9 (odd no.) 5th term 
"""
```

```python
import matplotlib.pyplot as plt

l= [25,46,57,25,45,13,29,39,46,99,22,12,14,16,55,6,46,52,55,78]
plt.boxplot(l)
plt.show()
```
