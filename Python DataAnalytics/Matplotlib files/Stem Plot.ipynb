```python
#Stem plots are valuable tools for visualizing individual data points, identifying trends,
#and comparing datasets. 
#Eg: Financial Data: Visualizing daily stock prices or other financial metrics 
#   to identify trends and outliers.
# Engineering: Analyzing signal data in electrical engineering or other measurements
#   in mechanical engineering.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = [10,40,20,40,20,40,20,40,60,70,50,40]
#plt.stem(x)
plt.stem(x,linefmt="--",markerfmt="D",bottom=10)
#bottom=0 default
#bottom=10 , y-axis starts with bottom value 10
#plt.stem(x,linefmt="--",markerfmt="D",bottom=10,orientation="horizontal")
plt.show()
```
