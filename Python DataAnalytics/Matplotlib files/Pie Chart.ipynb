```python
import matplotlib.pyplot as plt

brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
x = [22,35,30,3,10]
c = ["red","Violet","blue","pink","orange"]
ex = [0,0,0,0,0.1] #explode Redmi
#plt.pie(x)
#plt.pie(x,labels = brands,colors=c)
#plt.pie(x,labels = brands,colors=c,explode=ex) #Redmi piece comes out
#plt.pie(x,labels = brands,colors=c,explode=ex,shadow=True,autopct="%.2f") 
#autopct="%.2f"with decimal point 2
plt.pie(x,labels = brands,colors=c,explode=ex,shadow=True,autopct="%.2f",startangle=180)
# start angle=180 (It flips by 180 degrees)
#plt.savefig("pie1.png")
plt.savefig("pie2.png",facecolor='black',pad_inches=0.9,bbox_inches="tight")
#pad_inches - extra length added outside the pie chart to frame edge length
#bbox_inches - reduces the pic edge length to only sufficient frame
plt.show()
```
