```python
# Example form web
import seaborn as sns
import pandas as pd
import numpy as np

# Example data
np.random.seed(10)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=300),
    'Value': np.random.normal(loc=0, scale=1, size=300)
})

# Creating the violin plot
sns.violinplot(x='Category', y='Value', data=data)

# Adding titles and labels
plt.title('Violin Plot of Values by Category')
plt.xlabel('Category')
plt.ylabel('Value')

# Display the plot
plt.show()
```
