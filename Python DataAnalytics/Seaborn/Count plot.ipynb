```python
#Counts plot- It's used to plot no. of counts of any column value. 
# We use only x-axis here, no y-axis. Here hue is used as filter.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)

sns.countplot(data = data, x = "day")
plt.show()
```
