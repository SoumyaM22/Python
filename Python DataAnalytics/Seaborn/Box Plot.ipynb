```python
#Box plot - to know how many values are there nearby mean & outliers.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)

#sns.boxplot(data = data, x="tip")
# HEre in graph most of values lie btw 0 to 6, >6 values are outliers
#sns.boxplot(data = data, x="tip",y = "sex")
#sns.boxplot(data = data, x="sex",y = "tip",palette="GnBu")
sns.boxplot(data = data, x="sex",y = "tip",hue="day",palette="GnBu",fliersize=3)
#fliersize=3 to reduce size of outlier points
plt.show()
```
