```python
#Swarm plot - 
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
print(data)
#In swarm plot, data i.e points doesn't overlap between each other compared to strip plot
sns.swarmplot(data=data,x ="day",y="total_bill",hue="sex",dodge="True")
plt.show()
```

```python
#sns.stripplot(data=data,x ="day",y="total_bill")
#In strip plot, data i.e points overlaps between each other compared to swarm plot
sns.stripplot(data=data,x ="day",y="total_bill",jitter = "0.2",hue="sex",dodge="True")
plt.show()
```
