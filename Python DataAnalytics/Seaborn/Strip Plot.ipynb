```python
#strip plot- Similar to scatter plot i.e btw numerical & numerical values,
# here addition to scatter plot, 1 extra thing is that string also has data point values
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)

#sns.stripplot(data = data, x = "day", y = "total_bill",hue="sex")
sns.stripplot(data = data, x = "day", y = "total_bill",hue="sex",dodge=True,jitter=0.1)
#dodge=True is To get separate male, female values
#jitter=0.4, to move points in graph away
#jitter=0.1, to move points in graph close
#sns.scatterplot(data = data, x = "day", y = "total_bill") - only single dots
#sns.scatterplot(data = data, x = "tip", y = "total_bill")
plt.show()
```
