```python
# For histplot- only x-axis, count on other axis
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
#sns.histplot(data,x="day")
sns.histplot(data,x="day",hue="sex",kde="True")
#here kde means kluster density estimation - it helps us to know how frequently each value is being occured
plt.show()
```
