```python
#Violin plot - Used for numerical data like Age, Annual salary. 
# To find statistical data like where is mean, median value.
# To find frequency of data. We use only x-axis & no y-axis

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)

sns.violinplot(data = data,x = "total_bill")
plt.show()
# here frequency is more at 15 (which can identified as mean) 
# & median (white dot) is around 18. Black lin with White dot also considered as Bar plot
```
